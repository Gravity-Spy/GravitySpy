## Gravitational Wave Data

Let us start by taking some of the data from the detector and plotting it. To do this, we will be using GWpy, a collaboration-driven Python package that we imported above. GWpy methods will greatly simplify our manipulation of the time-domain and frequency-domain data coming from the LIGO and VIRGO interferometers.

```python
%matplotlib inline

# Pick the start and stop times and a channel name from which to read the data
eventTime   = 1126259462.3910
startTime   = eventTime - 32
stopTime    = eventTime + 32
channelName = 'L1:GDS-CALIB_STRAIN'
detector    = 'L'
frameType   = 'L1_HOFT_C00'

# Read in the data

NSDF = 0
if NSDF:
    data = TimeSeries.fetch(channelName,startTime,stopTime)
else:
    connection = datafind.GWDataFindHTTPConnection()
    cache = connection.find_frame_urls(detector, frameType, startTime, stopTime, urltype='file')
    data = TimeSeries.read(cache,channelName, format='gwf',start=startTime,end=stopTime)


```

We may want to sample the data at a particular frequency (the rate of samples per second, in units of Hertz). Equivalently, we may instead want to specify a time 'dt' between successive samples (in units of seconds). Here, we will choose to specify a sampling frequency (in Hertz) and we will check to make sure that our data is sampled at the desired frequency.

```python
# resample data if necessary
sampleFrequency = 4096
if data.sample_rate.decompose().value != sampleFrequency:
    data = data.resample(sampleFrequency)
```

Now that we have the time-domain data in the form of a TimeSeries object (which is just a time-domain data array), we want to be able to visualize this data. The plot should have the gravitational-wave strain amplitude (the magnitude of the length difference measured by the interferometers, corresponding to the strength of the gravitational-wave) on the y-axis, as a function of time on the x-axis.


```python
plot = data.plot()
plot.set_title('Timeseries')
plot.set_ylabel('Gravitational-wave strain amplitude')
plot.show()
```

This visualization is not very helpful: it is nearly impossible to pick out a signal from this data. Perhaps, we can look at the spectrogram of the data above. A spectrogram is a plot of frequency versus time, with strain amplitude represented as different colors (in the visualization below, higher amplitudes are yellow and lower amplitudes are blue). Essentially, this would require performing a fast fourier transform (FFT) on the TimeSeries data, in order to visualize it on what can also be called a "Time-Frequency Map." This is shown in [GWpy Spectrogram Example](https://gwpy.github.io/docs/v0.1/examples/spectrogram/plot.html). We will explore the concept of FFTs in more detail later on.

We will use 2 second long FFT with a 50 percent (1 second) overlap between segments. In addition we will use a stride of 5. Overlap is defined as the number of seconds between FFTs and stride is the number of seconds in the power spectral denisty (PSD), which represents one column of the spectrogram. 

The Spectrogram method first splits the TimeSeries data into different bins of equal time interval. The TimeSeries data is broken into chunks defined by the following time interval: [t - overlap/2, t + stride + overlap/2]. In other words, each time-bin is calculated using stride+overlap seconds of data. For each time-bin, a PSD value is calculated:  the PSD is essentially the power of the signal as a function of frequency, per unit frequency. It then plots a 3-dimensional (color--which represents power--being the third dimension) graph of time, frequency and power. 

Lets see this in action.

```python
specgram = data.spectrogram(5, fftlength=2, overlap=1) ** (1/2.)
plot = specgram.plot(norm='log', vmin=1e-23, vmax=1e-19)
plot.set_yscale('log')
plot.set_ylim(40, 2048)
plot.add_colorbar(
    label=r'Gravitational-wave amplitude [strain/$\sqrt{\mathrm{Hz}}$]',cmap='viridis')
```

It appears, from both the plot of the TimeSeries data and the spectrogram of the data, that no important discerning features can be achieved from looking only at the raw data. From the spectogram above it is clear that certain frequencies dominate over others, but why? To understand this phenomenon, lets create an Amplitude Spectral Density (ASD) graph, which essentially shows the sensitivity of the detector over different frequencies. To plot this ASD graph, we can use the convenient asd() method, which takes as arguments the FFT length, and overlap between FFT's. In our case, as mentioned above, we are passing an FFT length of 2 seconds, and overlap time of 1 second.

```python
asd = data.asd(2, 1)
```

Lets plot the ASD with the specified limits on the x and y axes

```python
plot = asd.plot(color='#ee0000', label='LIGO-Hanford')
ax = plot.gca()
ax.set_xlim(8, 2048)
ax.set_ylim(1e-24, 1e-20)
ax.set_ylabel(r'ASD [strain/\rtHz]')
plot.show()
```

So what does this ASD graph exactly tell us? The spikes in the graph correspond to the frequencies the detector is more sensitive to. If we look at the spectogram above, the frequencies where we see noise correspond to these spikes in our ASD graph. 

It looks like most of our noise is from loud lines in the data, at 500, 1000 and 1500 Hz. These lines correspond to Violin Mode Harmonics, a source of noise that arises from resonances of the wires that suspend the mirrors in the interferometer. We agains see similar spectral line at 60 Hz, which corresponds to the Powerline and its harmonics, a source of noise originating from the AC electrical supply. Because we are able to trace these visual features to their sources (and can therefore be sure that they represent noise as opposed to anything relevant to our search for a gravitational-wave signal), our next goal is to remove them through data processing. 

There are two ways by which we will get rid of excess noise in the data. A bandpass filter, and notching. A bandpass filter only allows data within a specified frequency range to pass through it. First, we will create a bandpass filter, to remove both low and high frequency components, leaving us with the frequency range of interest. We will implement this by using the gwpy.signal.filter_design module, which has a method called bandpass.

```python
bandpass = data.bandpass(50, 290, filtfilt=True)
plot = bandpass.plot(color='#ee0000')
ax = plot.gca()
ax.set_xlim(startTime,stopTime)
ax.set_ylim(-1.5e-21, 1.5e-21)
from gwpy.plotter import TimeSeriesPlot
plot.show()
```

Even with this bandpass filter there is not much information we can really grab from this plot. Lets 'notch' out the frequencies that were giving us trouble: the powerline harmonics at 60, 120 and 180 hz. We don't have to worry about the violin mode harmonic frequencies at 500, 1000 and 1500 because our bandpass filter took care of those already.

```python
notch = bandpass.notch(60).notch(120).notch(180)
plot = notch.plot(color='#ee0000')
ax = plot.gca()
ax.set_xlim(startTime,stopTime)
ax.set_ylim(-1.0e-21, 1.0e-21)
plot.show()
```

Finally, at around 16.5 seconds we can see a distinguishable spike. Lets zoom in and see what is going on.

```python
plot = notch.plot(color='#ee0000')
ax = plot.gca()
ax.set_xlim(eventTime - 1.0, eventTime + 1.0)
ax.set_ylim(-1.0e-21, 1.0e-21)
plot.refresh()
```

## The Concept of Match-Filtering

It is clear that we must come up with a good way to manipulate the data such that we can pull out signal from the data. One can imagine doing a match-filter to accomplish this task. Essentially, we simulate what the signal should look like and take the inner produce of it with your data. If your data and simulated signal are a good match then the numerical value that results should be high! Mathematically, this would look as follows.

$$
\tilde{d}(f)\cdot\tilde{h}_T(f) \equiv \int \tilde{d}(f) \tilde{h}_T(f)^{\star} df
$$

where $\tilde{h}_T(f)$ is your simulated signal

## Unmodeled Match Filtering
The overlap itself adds a weighting to each frequency component of the waveforms indicating how "sensitive" the detector is to that frequency. The sensitvity function is the power spectral density $S(f)$. So, the overlap is proportional to:

$$
\left(\tilde{h}_1(f)|\tilde{h}_2(f)\right) \propto 4 \int_0^{\infty} \frac{\tilde{h}_1(f) \tilde{h}_2(f)^{\star}}{S(f)} df
$$

Thus, we come up with the terminalogy of Signal (numerator: inner product) to Noise (denominator: average characterization of the noise). This value we call the SNR.

For our purposes, the process implemented above does not work because we are attempting to extract features from an unmodeled signal. We need to find a way to adapt this method to suit our situation.

From the Q-scan paper: "In general, a search algorithm for [unmodeled] gravitational-wave projects the data under test onto a basis constructed to span the space of plausible bursts"

So, what does that mean, exactly? Basically, "unmodeled" does not mean "unknown". We have reasonable expectations for what the signal should look like, at least roughly.

We define a number of parameters, such as central time, central frequency, duration, and bandwidth, which characterize the structure of bursts that would exist within the constraints of a time frequency spectrogram like the one plotted at the beginning of the iPython notebook.

*THIS* is our "simulated signal" so to speak. Therefore, we can imagine tweaking our beautiful match-filter equation from above:

$$
\int_0^{\infty} \frac{\tilde{d}(f) \tilde{h}(f_c,t_c,dur,band)^{\star}}{S(f)} df
$$

So, if we got lucky and exactly parameterized the characteristics of the burst signals, this equation would become:

$$
\int_0^{\infty} \frac{\tilde{d}^2(f)}{S(f)} df
$$

We know what we need to do! We have to attempt to parameterize the burst signal in the best possible way, because, if we do not, the parameterization encompasses either too little signal or too much noise, resulting in a loss in the measured signal-to-noise ratio. 

So what naturally characterizes burst signals? Their Q!

Wait a second, what is "their Q"?

The dimensionless quality factor Q for unmodeled GWs is the ratio of center frequency to bandwidth:

$$
Q = \frac{f_c}{band}
$$

In other words, Q is essentially a measure of the signal's aspect ratio in the time-frequency plane and is roughly equal to the number of oscillations of the time-domain waveform over the duration of the burst.

Since bursts are naturally characterized by their Q, we therefore choose to tile the timefrequency plane by selecting a measurement basis which directly targets bursts within a finite range of Q. 

For each burst, we can then perform an FFT (Fast Fourier Transform) for each frequency value within the bursts Q range, stepping logarithmically from one frequency value to the next. Once we have done this, we combine the spectrograms resulting from the FFTs for each frequency value. We include high intensity regions from each FFT, selecting the highest intensity signal for regions in which high intensity aspects of multiple FFTs overlap. 

This naturally leads to a logarithmic tiling of the timefrequency plane in which individual measurement pixels are well-localized signals, all with the same Q.

Thus, the Q transform at a specific frequency is obtained by a standard Fourier transform
of the original time series, a shift in frequency, multiplication by the appropriate frequency
domain window function, and an inverse Fourier transform.
